**ENCAPSULATION in Python**

Encapsulation is fundamental principles of object-oriented programming in python that focuses on bundling data and methods together within a class, and controlling the access to that data from outside the class.

```python
class OOPinPython:
    def __init__(self, concept, difficulty):
    # Public member variable
        self.concept = concept
    # Private member variable
        self.__difficulty = difficulty
    # Public method
    def get_concept(self):
        return self.concept
    #Private method
    def __get_difficulty(self):
        return self.__difficulty

    def display_info(self):
        print(f"Concept: {self.concept}")
        print(f"Difficulty: {self.__difficulty}")
        print("---------------")




OOPinPython_obj2 = OOPinPython("Encapsulation","Intermediate")
OOPinPython_obj2.display_info()

try:
    print(OOPinPython_obj2.concept)  # Attempting to access public member outside the class
except AttributeError:
    print("\nError: Private method cannot be accessed outside the class!!")

try:
    print(OOPinPython_obj2.get_concept())  # Attempting to access public method outside the class
except AttributeError:
    print("\nError: Private method cannot be accessed outside the class!!")

try:
    print(OOPinPython_obj2.__difficulty)  # Attempting to access private member outside the class
except AttributeError:
    print("\nError: Private member cannot be accessed outside the class!!")


try:
    print(OOPinPython.__get_difficulty)  # Attempting to access private member outside the class
except AttributeError:
    print("\nError: Private method cannot be accessed outside the class!!")

```

**INHERITANCE in Python**

Inheritance allows us to define a class that inherits all the methods and properties from another class.

Parent class is the class being inherited from, also called base class.

Child class is the class that inherits from another class, also called derived class.

```python
class OOPConcept:
    def __init__(self, concept, difficulty):
        self._concept = concept
        self._difficulty = difficulty

    def display_info(self):
        print(f"Concept: {self._concept}")
        print(f"Difficulty: {self._difficulty}")
        print("---------------")


class Polymorphism(OOPConcept):
    def __init__(self, concept, difficulty, usage):
        super().__init__(concept, difficulty)
        self._usage = usage

    def display_info(self):
        super().display_info()
        print(f"Usage of Polymorphism in Python: {self._usage}")
        print("---------------")


class Encapsulation(OOPConcept):
    def __init__(self, concept, difficulty, importance):
        super().__init__(concept, difficulty)
        self._importance = importance

    def display_info(self):
        super().display_info()
        print(f"Importance of Encapsulation in Python: {self._importance}")
        print("---------------")


class Inheritance(OOPConcept):
    def __init__(self, concept, difficulty, advantages):
        super().__init__(concept, difficulty)
        self._advantages = advantages

    def display_info(self):
        super().display_info()
        print(f"Advantages of Inheritance in Python: {self._advantages}")
        print("---------------")

OOPConcept_obj = OOPConcept("OOP CONCEPTS", "Beginner")
OOPConcept_obj.display_info()

polymorphism = Polymorphism("Polymorphism", "Intermediate", "Can have many forms")
polymorphism.display_info()

encapsulation = Encapsulation("Encapsulation", "Beginner", "Provides data protection")
encapsulation.display_info()

inheritance = Inheritance("Inheritance", "Advanced", "Code reusability and extensibility")
inheritance.display_info()
```

**DATA ABSTRACTION in Python**

Data Abstraction in python is defined as a process of handling complexity by hiding unnecessary information from the user. An abstract method is a method that is declared, but does not contain implementation. An abstract method in a base class identifies the functionality that should be implemented by all its subclasses.


```python
from abc import ABC, abstractmethod

class OOPConcept(ABC):
    def __init__(self, concept, difficulty):
        self._concept = concept
        self._difficulty = difficulty

    @abstractmethod
    def display_info(self):
        pass


class Polymorphism(OOPConcept):
    def __init__(self, concept, difficulty, usage):
        super().__init__(concept, difficulty)
        self._usage = usage

    def display_info(self):
        print(f"Concept: {self._concept}")
        print(f"Difficulty: {self._difficulty}")
        print(f"Usage of Polymorphism in Python: {self._usage}")
        print("---------------")


class Encapsulation(OOPConcept):
    def __init__(self, concept, difficulty, importance):
        super().__init__(concept, difficulty)
        self._importance = importance

    def display_info(self):
        print(f"Concept: {self._concept}")
        print(f"Difficulty: {self._difficulty}")
        print(f"Importance of Encapsulation in Python: {self._importance}")
        print("---------------")


class Inheritance(OOPConcept):
    def __init__(self, concept, difficulty, advantages):
        super().__init__(concept, difficulty)
        self._advantages = advantages

    def display_info(self):
        print(f"Concept: {self._concept}")
        print(f"Difficulty: {self._difficulty}")
        print(f"Advantages of Inheritance in Python: {self._advantages}")
        print("---------------")


polymorphism = Polymorphism("Polymorphism", "Intermediate", "Can have many forms")
polymorphism.display_info()

encapsulation = Encapsulation("Encapsulation", "Beginner", "Provides data protection")
encapsulation.display_info()

inheritance = Inheritance("Inheritance", "Advanced", "Code reusability and extensibility")
inheritance.display_info()
```
